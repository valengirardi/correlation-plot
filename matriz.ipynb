```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo
file_path = "/Users/valentinagirardi/Downloads/bioinformatica-todaslascarpetas/reactores2023/PCA.XLSX"
datos = pd.read_excel(file_path)

# Eliminar las columnas 'type-of-reactor' y 'sample-id'
datos = datos.drop(columns=['type-of-reactor', 'sample-id'])

# Crear el grfico de correlacin
correlation_matrix = datos.corr()

# Personalizar el estilo del grfico
plt.figure(figsize=(30, 10))

# Generar el mapa de calor
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Aadir ttulo
plt.title('Matriz de Correlacin')

# Mostrar el grfico
plt.show()

```
